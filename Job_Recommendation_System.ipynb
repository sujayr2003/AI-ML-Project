```python
#After Removal of duplcate and null values from the job_description column we have to then
#remove all the links and punctuations from the job_description as it is of now relevance
#to our model
from bs4 import BeautifulSoup
import re
def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()
def remove_punctuation(text):
    return re.sub(r'[^\w\s\n]', '', text)
df['job_description'] = df['job_description'].apply(remove_html_tags)
df['job_description'] = df['job_description'].apply(remove_punctuation)
df.drop_duplicates(subset=['job_description'],inplace=True)
df['job_description'] = df['job_description'].str.replace('\n', '')
df.drop_duplicates(subset=['job_description'],inplace=True)
df['job_description'].value_counts()
```

```python
#importing nltk (Natural Language Toolkit) library and stopwords from nltk.corpus library
#to remove all stopwords like 'over','or' and etc from the column job_description
import nltk
from nltk.corpus import stopwords
# Load the stopwords
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    # Tokenize the text
    words = text.split()
    # Remove stop words
    filtered_words = [word for word in words if word.lower() not in stop_words]
    # Join the filtered words back into a sentence
    return ' '.join(filtered_words)
# Removing all stop words from the column job_description so our model will use only the most related words from job description column
df['job_description'] = df['job_description'].apply(remove_stopwords)
df
```

```python
#Recommendation based on candidates current/previous occupation
#First we have to import TfidVectorizer model from scikit-learn present in the module feature_extraction.text
from sklearn.feature_extraction.text import TfidfVectorizer
#Then we have to create an object of TFIDF Vectorizer to use the tdidf model
tfidfvec = TfidfVectorizer()
#Each job title in the 'job_title' column is tokenized into individual words or n-grams.
#The TF-IDF vectorizer constructs a vocabulary of all unique tokens (words or n-grams) observed in the 'job_title' column.
#This vocabulary is built based on the on the data from job_title column
#The TF-IDF vectorizer calculates the document frequency (DF) of each term in the vocabulary.
#DF is the number of documents (job titles) in which a term appears.
#The TF-IDF vectorizer computes the inverse document frequency (IDF) for each term in the vocabulary.
# IDF is calculated as the logarithm of the ratio of the total number of documents to the document frequency of each term.
#For each job title (document), the TF-IDF vectorizer computes the TF-IDF score for each term in the vocabulary.
#TF-IDF is calculated as the product of the term frequency (TF) and inverse document frequency (IDF) for each term in the document.
#The variable tfidf_jobtitles stores the resulting TF-IDF matrix. It is a  matrix representation with 0's , 1's and fractions
#where each row corresponds to a job title, and each column corresponds to a term in the vocabulary.
#The values in the matrix represent the TF-IDF scores of each term in each job title.
#The tdidf scores represent the importance of each term/word/n-gram within a documnet relative t the entire collection of documents
#The final matrix recevied as output will treated as a vector
tfidf_jobtitles = tfidfvec.fit_transform(df['job_title'])
print
#From the module sklearn.metrics.pairwise present in scikit learn library we are importing cosine similarity function
from sklearn.metrics.pairwise import cosine_similarity
#To make the recommendation we will have compute the cosine similarity so that we can give the recommendation to the candidate based on the job titles
#having the highest similairty scores.
#To compute this cosine similairity of job_titles.
#The dot product of the two vectors(both of which are the tdidf scores matrix) is divided by the product of their magnitudes which will then give the  cosine similarity
#The cosine similarity is a measure of similarity between two vectors and ranges from -1 to 1.
#where 1 indicates that the vectors are identical in direction, 0 indicates no similarity, and -1 indicates opposite directions.
#The resultant of this similarity variable is a symmentric matrix containing the similairty scores of the job_titles from job_title column based on the tdidf scores
#where each element contains the cosine similarity between each document from vector 1 and vector 2
similarity = cosine_similarity(tfidf_jobtitles, tfidf_jobtitles)
# Creating a pandas series containing all the indices and job_titles from the column job title
indices = pd.Series(df['job_title'])
#Creating a function to  give  recommendations to the candidate
#where the candidate only has to enter their job title and the function will take the cosine similarity by default
def recommendation(title, cosine_sim=similarity):
    try:
        #This line returns the job title given by the user that is similar to the one in the dataset
        index = indices[indices == title].index[0]
        #Now sorting the similairty scores in descending order in comparision to the index values computed above
        similarity_scores = pd.Series(similarity[index]).sort_values(ascending=False)
        #This line gives the top 25 job titles indices having highest similairty score
        top_25_jobs = list(similarity_scores.iloc[1:25].index)
        #This line gives the job titles associated with the indices computed above
        recommended_jobs = [list(df['job_title'])[i] for i in top_25_jobs]
        #Returning the job titles
        return recommended_jobs

    except:
        #Returning that no job title has ben found if the job title entered by the user is not in the dataset
        print("No Job Title is found")
```

```python
#Giving recommendations to the user
Input='C# Software Engineer'
recommendation(Input)
```

```python
# Import SequenceMatcher from difflib module
from difflib import SequenceMatcher

# Set input text (assumed to be provided elsewhere)
input_text = Input

# Get the output texts from the recommendation function
output_texts = recommendation(Input)

# Initialize an empty list to store ground truth
ground_truth = []

# Iterate over each output text
for output in output_texts:
    # Iterate over the characters in the input text
    for i in range(len(input_text)):
        # Check if the character in the input text is present in the output text
        if input_text[i] in output:
            # If a match is found, add the output text to the ground truth list
            ground_truth.append(output)

# Define the similarity threshold
similarity_threshold = 0.6

# Initialize counters for true positives, false positives, and false negatives
true_positives = 0
false_positives = 0

# Iterate over each output text
for output in output_texts:
    # Calculate the similarity between the input text and the output text using SequenceMatcher
    similarity = SequenceMatcher(None, input_text, output).ratio()

    # Check if the similarity meets the similarity threshold
    if similarity >= similarity_threshold:
        # Check if the output text is in the ground truth list
        if output in ground_truth:
            # If the output text is in the ground truth list, it's a true positive
            true_positives += 1
        else:
            # If the output text is not in the ground truth list, it's a false positive
            false_positives += 1

# Calculate the number of false negatives
false_negatives = len(ground_truth) - true_positives

# Calculate precision, recall, and F1 score, handling potential division by zero
precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
accuracy = true_positives / len(ground_truth) if len(ground_truth) > 0 else 0

# Print the computed metrics
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
print("Accuracy:", accuracy)
```
